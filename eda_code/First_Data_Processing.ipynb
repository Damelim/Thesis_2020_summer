3. action  

R2의 shottype ==2 | R2의 FoulType == 'shooting' & R3의 FreeThrowNum에 'of 2'라는 문구가 있을 때 -> 2

R2의 shottype ==3 | R2의 FoulType == 'shooting' & R3의 FreeThrowNum에 'of 3'라는 문구가 있을 때 -> 3

(R2의 FoulType =='personal' or 'loose ball' -> common foul : 빼면 좋겠다.)

R2의 TurnoverType =='common_tov' -> common tov

R2의 TurnoverType =='steal' -> steal


```python
if (sample_action.ShotType[1] ==2 or (sample_action.FoulType[1] == 'shooting' and sample_action.FreeThrowNum.str.contains('of 2',na=False)[3] == True)):
    modeling_data.action[0] = '2pt'

elif (sample_action.ShotType[1] ==3 or (sample_action.FoulType[1] == 'shooting' and sample_action.FreeThrowNum.str.contains('of 3',na=False)[3] == True)):
    modeling_data.action[0] = '3pt'
    
elif sample_action.TurnoverType[1] =='common_tov':
    modeling_data.action[0] = 'common_tov'
    
elif sample_action.TurnoverType[1] =='steal':
    modeling_data.action[0] = 'steal'
    
elif sample_action.FoulType[1].isin(['personal','loose_ball']):
    modeling_data.action[0] = 'common_foul'
```

4. action_player

```python
sample_action.head()

if sample_action.Shooter.notnull()[1] == True :
    modeling_data.action_player[0] = sample_action.Shooter[1]
elif sample_action.Fouled.notnull()[1] == True :
    modeling_data.action_player[0] = sample_action.Fouled[1]
elif sample_action.TurnoverPlayer.notnull()[1] == True :
    modeling_data.action_player[0] = sample_action.TurnoverPlayer[1]
```

5. possession time

6. result

```python
if sample_action.AwayPlay.notnull()[2] == True:
    modeling_data.result[0] = 'ORB'
elif sample_action.ShotOutcome[1] == 1:
    modeling_data.result[0] = 'made'
elif sample_action.AwayPlay.notnull()[2] == False:
    modeling_data.result[0] = 'DRB'
else :
    modeling_data.result[0] = 'SF'
```

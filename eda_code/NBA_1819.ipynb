```python
import pandas as pd
import numpy as np
```

```python
data_1819 = pd.read_csv('C:/2020_thesis_originaldata/NBA_PBP_2018_2019.csv')
data_1819.head()
```

```python
data_1819.drop(columns = ['Location','Date','Time','WinningTeam','JumpballAwayPlayer','JumpballHomePlayer','JumpballPoss','TimeoutTeam'], inplace= True)
```

```python
data_1819.drop(columns = ['ShotDist','Assister'], inplace= True)
```

```python
data_1819 = data_1819[['URL','GameType','HomeTeam','AwayTeam','Quarter','SecLeft','HomePlay','AwayPlay','HomeScore','AwayScore','Shooter', 'ShotType',
       'ShotOutcome', 'Blocker', 'FoulType', 'Fouler',
       'Fouled', 'Rebounder', 'ReboundType', 'ViolationPlayer',
       'ViolationType', 'FreeThrowShooter', 'FreeThrowOutcome', 'FreeThrowNum',
       'EnterGame', 'LeaveGame', 'TurnoverPlayer', 'TurnoverType',
       'TurnoverCause', 'TurnoverCauser']]
```

```python
len(data_1819.columns)
```

```python
data_1819['URL'] = data_1819.URL.str[15:23]
data_1819
```

```python
data_1819.GameType[data_1819.GameType == 'regular'] = 'reg'
data_1819.GameType[data_1819.GameType == 'playoff'] = 'plo'
```

```python
data_1819['ShotType'] = data_1819.ShotType.str[0]
data_1819[['ShotType']]
data_1819['ShotType'] = data_1819.ShotType.astype('float')
data_1819[['ShotType']]
```

```python
data_1819.ShotOutcome[data_1819.ShotOutcome == 'make'] = 1

data_1819.ShotOutcome[data_1819.ShotOutcome == 'miss'] = 0

data_1819[['ShotOutcome']]
```

```python
data_1819.FreeThrowOutcome[data_1819.FreeThrowOutcome == 'make'] = 1

data_1819.FreeThrowOutcome[data_1819.FreeThrowOutcome == 'miss'] = 0

data_1819[['FreeThrowOutcome']]
```

```python
data_1819 = data_1819[data_1819.Quarter.isin([1,2,3,4])]

data_1819.shape
```

```python
data_1819 = data_1819[data_1819.ViolationType.isnull()]

data_1819.ViolationType.value_counts()
```

```python
data_1819.drop(columns= ['ViolationPlayer','ViolationType'],inplace=True)

data_1819.shape
```

```python
data_1819.FreeThrowNum = data_1819.FreeThrowNum[data_1819.FreeThrowNum != 'technical']

data_1819
```

```python
data_1819['TurnoverType'][data_1819['TurnoverCause'] =='steal'] = 'steal'
```

```python
common_tov_array = data_1819[data_1819['TurnoverType'].notnull()].TurnoverType.unique()
common_tov_array = common_tov_array[common_tov_array != 'steal']
```

```python
data_1819.TurnoverType[data_1819.TurnoverType.isin(common_tov_array)] = 'common_tov'
data_1819.TurnoverType.value_counts()
```

```python
data_1819.drop(columns = ['TurnoverCause','TurnoverCauser'], inplace=True)
data_1819
```

```python
data_1819.columns
```

```python
person_col_list = ['Shooter','Blocker','Fouler','Fouled','Rebounder','FreeThrowShooter','TurnoverPlayer']
```

```python
data_1819[person_col_list[0]].str[:-5]
```

```python
data_1819[person_col_list]
```

```python
for i in person_col_list:
    data_1819[i] = data_1819[i].str[:-5]
```

```python
#df[df['A'].str.contains("hello")]
data_1819 = data_1819[~data_1819['HomePlay'].str.contains('timeout',na=False)]
data_1819 = data_1819[~data_1819['AwayPlay'].str.contains('timeout',na=False)]
```

```python
#df[df['A'].str.contains("hello")]
data_1819 = data_1819[~data_1819['HomePlay'].str.contains('technical',na=False)]
data_1819 = data_1819[~data_1819['AwayPlay'].str.contains('technical',na=False)]
```

```python
data_1819.reset_index(inplace=True)
data_1819.drop(columns = ['index'],inplace=True)
data_1819
```

```python
data_1819reg = data_1819[data_1819.GameType == 'reg']
data_1819plo = data_1819[data_1819.GameType == 'plo']
data_1819plo.reset_index(inplace=True)
data_1819plo.drop(columns = ['index'], inplace=True)
data_1819reg.shape[0] + data_1819plo.shape[0] == data_1819.shape[0]
```

```python
data_1819reg.drop(columns = ['GameType'], inplace = True)
data_1819plo.drop(columns = ['GameType'], inplace = True)
```

```python
data_1819reg.to_csv('C:/2020_thesis_repo/original_data/data_1819reg.csv')
data_1819plo.to_csv('C:/2020_thesis_repo/original_data/data_1819plo.csv')
```
